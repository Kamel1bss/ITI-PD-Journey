```python
import numpy as np
from numpy.random import randn
```

```python
np.random.seed(150000) # return the same random array each time.
df = pd.DataFrame(index=['X', 'Y', 'Z'], columns=['C1', 'C2', 'C3'], data = randn(3, 3) )
df
```

```python
# Printing columns
df[ ['C1', 'C3'] ]
```

```python
df['Sum'] = df['C1'] + df['C2'] + df['C3']
```

```python
df
```

```python
# Printing rows
df.loc[['X', 'Y']]
```

```python
df.loc[ ['X', 'Y'], ['C2', 'C3'] ]
```

```python
# Printing cells
```

```python
df.loc['X', 'C2']
```

```python
# Droping columns[axis → 1] 
df.drop('Sum', axis=1, inplace=True)
```

```python
# Droping rows[axis → 0]
df.drop('Z', axis=0, inplace=True)
```

```python
df
```

```python
# Filtering the data
df_2 = df[ (df['C3'] >= 0) | (df['C1'] >= .5)].loc['Y']
```
