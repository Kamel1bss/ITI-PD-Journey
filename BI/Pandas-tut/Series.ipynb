```python
import pandas as pd
```

```python
headers = ['Name', 'Age', 'Major']
values = ['Kamel', 23, 'CS']

# Series is a (key, value) pair like a dictionary
# index → keys
# data → values
my_series_1  = pd.Series(data=values, index=headers)
```

```python
my_series_1['Name']
```

```python
my_series_2 = pd.Series(values)
```
